```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df =pd.read_csv('Dados/reported.csv')
```

```python
df.head()
```

```python
df.fillna(0, inplace= True)
```

```python
df.head()
```

```python
plt.plot(df['Year'], df['crimes.total'], '-r')
plt.plot(df['Year'], df['crimes.person'], '-b')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(df['Year'], df['crimes.total'], '-r')
ax.plot(df['Year'], df['crimes.person'], '-b')
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(df['Year'], df['crimes.total'], '-r')
ax.plot(df['Year'], df['crimes.person'], '-b')
ax.legend(loc='upper right')
```

```python
fig, ax = plt.subplots()
ax.plot(df['Year'], df['crimes.total'], '-r')
ax.plot(df['Year'], df['crimes.person'], '-b')
ax.legend(loc='lower left')
```

```python
fig, ax = plt.subplots()
ax.plot(df['Year'], df['crimes.total'], '-r')
ax.plot(df['Year'], df['crimes.person'], '-b')
ax.legend(loc='upper left')
ax.set_ylabel('Quantidade')
ax.set_xlabel('Ano')
ax.set_title('Crimes: Total X Person')
```

```python
fig, ax = plt.subplots()
ax.plot(df['Year'], df['crimes.total'], '-r')
ax.plot(df['Year'], df['crimes.person'], '-b')
ax.legend(loc='upper left')
ax.set_ylabel('Quantidade')
ax.set_xlabel('Ano')
ax.set_title('Crimes: Total X Person')
ax.set_xlim([df['Year'].min(),df['Year'].max()])
ax.plot()
```
