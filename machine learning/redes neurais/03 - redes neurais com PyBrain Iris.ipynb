```python
import numpy as np
from pybrain.datasets import SupervisedDataSet
from pybrain.tools.shortcuts import buildNetwork
from pybrain.supervised import BackpropTrainer
```

```python
entradas = np.genfromtxt('../Dados/iris2.data', delimiter=',', usecols=(0,1,2,3))
saidas = np.genfromtxt('../Dados/iris2.data', delimiter=',', usecols=(4))
print(len(entradas))
print(len(saidas))
```

```python
'''
Iris-setosa: 0
Iris-versicolor: 1
Iris-virginica: 2
'''

print(entradas[:5])
print(saidas[:5])
```

```python
#uso de 70% da amostragem para treino
entradas_treino = np.concatenate((entradas[0:35], entradas[50:85], entradas[100:135]))
saidas_treino = np.concatenate((saidas[0:35], saidas[50:85], saidas[100:135]))
```

```python
#Uso de 30% para teste
entradas_teste = np.concatenate((entradas[35:50], entradas[85:100], entradas[135:]))
saidas_teste = np.concatenate((saidas[35:50], saidas[85:100], saidas[135:]))
```

```python
treinar = SupervisedDataSet(4,1)
for i in range(len(entradas_treino)):
    treinar.addSample(entradas_treino[i], saidas_treino[i])
print(len(treinar))
print(treinar.indim)
print(treinar.outdim)
    
```

```python
#construindo rede
rede = buildNetwork(treinar.indim, 2, treinar.outdim, bias=True)
treino = BackpropTrainer(rede, treinar, learningrate=0.01, momentum=0.3)

#trenar rede
for epoca in range(1000):
    treino.train()
    
#testando rede
teste = SupervisedDataSet(4,1)
for i in range(len(entradas_teste)):
    teste.addSample(entradas_teste[i],saidas_teste[i])
treino.testOnData(teste, verbose=True)
```

```python
#construindo rede
#alterando momentum, learnigrate e ciclos de epoca
rede = buildNetwork(treinar.indim, 2, treinar.outdim, bias=True)
treino = BackpropTrainer(rede, treinar, learningrate=0.005, momentum=0.5)

#trenar rede
for epoca in range(2000):
    treino.train()
    
#testando rede
teste = SupervisedDataSet(4,1)
for i in range(len(entradas_teste)):
    teste.addSample(entradas_teste[i],saidas_teste[i])
treino.testOnData(teste, verbose=True)
```
