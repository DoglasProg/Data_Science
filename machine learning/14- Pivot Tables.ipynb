```python
import pandas as pd
import numpy as np
```

```python
primary_df = pd.read_csv('Dados/primary-results.csv')
```

```python
primary_df[(primary_df['candidate'] == 'Bernie Sanders')&
          (primary_df['state'] == 'Alabama')]
```

```python
primary_df['candidate'].unique()
```

```python
#organiza relaçoes de dados
pd.pivot_table(primary_df, index=['state', 'party', 'candidate'])
```

```python
#coloca como dado de referencia os votos
pd.pivot_table(primary_df, index=['state', 'party', 'candidate'], values='votes')
```

```python
#coloca como dado de referencia os votos, aplica soma de votos
pd.pivot_table(primary_df, index=['state', 'party', 'candidate'], values='votes'
              ,aggfunc={'votes' : np.sum})
```

```python
pd.pivot_table(primary_df, index=['state', 'party', 'candidate'], values='fraction_votes'
              ,aggfunc={'fraction_votes' : np.sum})
```

```python
gr = primary_df.groupby(['county', 'party'])['votes'].rank(ascending=False)
primary_df['rank'] = gr
```

```python
primary_df[primary_df['county']== 'Los Angeles']
```

```python
primary_df_groupby = primary_df.groupby(['state', 'party', 'candidate']).sum()
del primary_df_groupby['fips']
del primary_df_groupby['fraction_votes']
primary_df_groupby.reset_index(inplace = True)
primary_df_groupby.head(10)
```

```python
primary_df_groupby[primary_df_groupby['rank'] < 5]
```

```python
primary_df_groupby['rank'] = primary_df_groupby.groupby(
    ['state', 'party'])['votes'].rank(ascending=False)
```

```python
primary_df_groupby.head(7)
```

```python
pd.pivot_table(primary_df_groupby, index=['state', 'party', 'candidate'], 
              values=['rank', 'votes'])
```
