```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
import numpy as np
```

```python
salario = np.array([100,200,300,500,50])
salario
```

```python
plt.plot(salario)
```

```python
plt.plot(salario, c='Black') #cor da linha
```

```python
plt.plot(salario, c='Black', ls= '--')
```

```python
plt.plot(salario, c='Red', ls='--')
```

```python
plt.plot(salario, c='Red', ls='--', marker='s')
```

```python
plt.plot(salario, c='Black', ls='--', marker='s', ms=10)
```

```python
plt.plot(salario, c='red', marker='s', ls='--', ms=8)
plt.show() #retira mensagem 
```

```python
sp1 = np.array([10,200,300,400])
sp2 = np.array([50,150,250, 350])
```

```python
plt.plot(sp1, c='red', marker='s', ls='--', ms=8, label='sps1')
plt.plot(sp2, c='Black', marker='s', ls='--', ms=8, label='sps2')
plt.show()
```

```python
plt.plot(sp1, c='red', marker='s', ls='--', ms=8, label='sps1')
plt.plot(sp2, c='Black', marker='s', ls='--', ms=8, label='sps2')
plt.legend()
plt.show()
```

```python
plt.plot(sp1, c='red', marker='s', ls='--', ms=8, label='sps1')
plt.plot(sp2, c='Black', marker='^', ls='--', ms=8, label='sps2')
plt.legend()
plt.show()
```

```python
plt.plot(sp1, c='red', marker='s', ls='--', ms=8, label='sps1')
plt.plot(sp2, c='Black', marker='^', ls='--', ms=8, label='sps2')
plt.legend(loc = 'upper right')
plt.show()
```

```python
sp3 = [100,50,500,350]
plt.plot(sp1, c='red', marker='s', ls='--', ms=8, label='sps1')
plt.plot(sp2, c='Black', marker='^', ls='--', ms=8, label='sps2')
plt.plot(sp3, c='Green', marker='o', ls='--', ms=8, label='sps3')
plt.legend(loc = 'upper left')
plt.show()
```
